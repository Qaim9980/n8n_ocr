```python
def fixed_sliding_window(arr, k):
    # Edge case: if array length is smaller than window size
    if len(arr) < k:
        print("Window size is larger than array length.")
        return 0

    # Step 1: Calculate sum of first window
    window_sum = sum(arr[:k])
    max_sum = window_sum

    print(f"Initial window ({arr[:k]}) -> Sum = {window_sum}")

    # Step 2: Slide the window
     for i in range(k, len(arr)):
        outgoing = arr[i - k]   # element leaving the window
        incoming = arr[i]       # element entering the window

        window_sum = window_sum - outgoing + incoming

        print(
            f"Window moved: -{outgoing} +{incoming} "
            f"-> Current Sum = {window_sum}"
        )

        # Update maximum sum
        if window_sum > max_sum:
            max_sum = window_sum
            print(f"New max found: {max_sum}")

    return max_sum


# ðŸ” Test the function
arr = [2, 1, 5, 1, 3, 2]
k = 3

result = fixed_sliding_window(arr, k)
print("\nMaximum sum of subarray of size", k, "is:", result)
```

```python

def variable_sliding_window(arr, target):
    left = 0
    current_sum = 0
    min_length = float('inf')

    print(f"Array: {arr}")
    print(f"Target Sum: {target}\n")

    for right in range(len(arr)):
        # Expand window
        current_sum += arr[right]
        print(f"Right moved to index {right}, added {arr[right]} â†’ current_sum = {current_sum}")

        # Shrink window while condition is satisfied
        while current_sum >= target:
            window_length = right - left + 1
            min_length = min(min_length, window_length)

            print(
                f"  Window [{left}:{right}] â†’ {arr[left:right+1]}, "
                f"sum = {current_sum}, length = {window_length}"
            )

            # Shrink from left
            current_sum -= arr[left]
            print(f"  Removing {arr[left]} from sum")
            left += 1
            print(f"  Left moved to index {left}, current_sum = {current_sum}\n")

    # Final result
    if min_length == float('inf'):
        print("No subarray found with sum >= target")
        return 0
    else:
        print(f"Minimum subarray length with sum >= {target} is {min_length}")
        return min_length


# ðŸ”¹ Example Usage
arr = [2, 3, 1, 2, 4, 3]
target = 7

result = variable_sliding_window(arr, target)
print("\nFinal Answer:", result)

```
